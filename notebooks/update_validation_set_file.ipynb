```python
import pandas as pd
from pathlib import Path
from typing import Union, TextIO

def parse_exposure_file_to_df(file: Union[str, Path, TextIO]) -> pd.DataFrame:
    """
    Parse a file where each line has:
        <short_exposure_file> <long_exposure_file> <ratio>

    Returns a DataFrame with columns:
        ['short', 'long', 'ratio']
    """
    if isinstance(file, (str, Path)):
        # file path
        df = pd.read_csv(
            file,
            sep=r"\s+",
            header=None,
            names=["short", "long", "ratio"],
            engine="python",
        )
    else:
        # already-open file-like object
        df = pd.read_csv(
            file,
            sep=r"\s+",
            header=None,
            names=["short", "long", "ratio"],
            engine="python",
        )

    # make sure ratio is integer
    df["ratio"] = df["ratio"].astype(int)
    return df
```

```python
df = parse_exposure_file_to_df("/home/david.weijiecai/computational_imaging/ExposureDiffusion/SID_Sony_15_paired.txt")
df["short"] = df.short.str.replace("ARW", "tiff")
```

```python
import pandas as pd
from pathlib import Path
from typing import Union

def save_exposure_df_to_txt(df: pd.DataFrame, path: Union[str, Path]) -> None:
    """
    Save a DataFrame with columns ['short', 'long', 'ratio'] to a txt file
    in the format:
        <short> <long> <ratio>

    Example line:
        20188_00_0.1s.tiff 20188_00_10s.ARW 100
    """
    # Ensure correct column order and cast ratio to int (or str if you prefer)
    out = df[["short", "long", "ratio"]].copy()
    # If ratio might be float, uncomment this:
    # out["ratio"] = out["ratio"].astype(int)

    out.to_csv(path, sep=" ", header=False, index=False)
```
